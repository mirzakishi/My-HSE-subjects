#### Localhost and Loopback IP adress(127.0.0.1)

Loopback is the routing of electronic signals, digital data streams back to their source without intentional processing or modification. It is primarily a means of testing the communications infrastructure. 
Loopback is the special virtual network(or virtual server) that the computer communicate with itself.

Used for:

1. Diagnostics
2. Troubleshooting
3. Connect to servers running on local machine

127.0.0.1 is the loopback Internet Protocol(IP) addresss also referred to as the "localhost". If our computer IP address(for example: 91.228.178.74) used for communicate with other comuters and networks, localhost(127.0.0.1) used for communication computer with itself.
